```python
#import Librabries
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

df=pd.read_csv('/content/Titanic Dataset (2).csv')
df.head()
```

```python
#missing values
print("..........missing values.......")
df.isnull().sum()
```

```python
#rremove missing values
df=df.dropna()
print("....after removing....")
df.isnull().sum()
```

```python
#select input features and target
inputs=df[['Pclass','Sex','Age','Fare']].copy()
target=df['Survived']

print("input features")
inputs.head()
target.head()
```

```python
#encode categorical data
print(".........label encode........")
# Re-initialize inputs to ensure 'Sex' column is in its original string format
inputs=df[['Pclass','Sex','Age','Fare']].copy()

le=LabelEncoder()
inputs['Sex']=le.fit_transform(inputs['Sex'])
print(inputs.head())

print("\nEncoder:")
print("Female=",le.transform(['female'])[0])
print("Male=",le.transform(['male'])[0])
```

```python
#split training and testing data
inputs_train,inputs_test,target_train,target_test=train_test_split(inputs,target,test_size=0.5,random_state=42)
print("training data shape:",inputs_train.shape)
print("testing data shape:",inputs_test.shape)
```

```python
#create decision tree model
model=DecisionTreeClassifier(criterion="gini")
#train model
model.fit(inputs_train,target_train)
#model accuracy
train_accuracy=model.score(inputs_train,target_train)
test_accuracy=model.score(inputs_test,target_test)
overall_accuracy=model.score(inputs,target)
print("train accuracy:",train_accuracy)
print("test accuracy:",test_accuracy)
print("overall accuracy:",overall_accuracy)
```

```python

# Predict a new passenger
# Format: [Pclass, Sex, Age, Fare]
new_passenger = [[1, 1, 58.0, 26.5500]]

prediction = model.predict(new_passenger)

print("------- New Passenger Prediction -------")

if prediction[0] == 1:
    print("Prediction: Survived")
else:
    print("Prediction: Did Not Survive")
```
