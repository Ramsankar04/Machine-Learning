```python
# IMPORT THE LIBRARIES

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# STEP 2: LOAD THE DATASET

df = pd.read_csv(r"C:\Users\user\OneDrive\Titanic-Dataset.csv")

print(df.head())

# Checking For Missing Values
print(df.isnull().sum())

# Removing Missing Values
df = df.dropna()

# STEP 3: REMOVE MISSING VALUES

print("\n-----AFTER REMOVING MISSING VALUES-----")
print(df.isnull().sum())

# STEP 4: SELECT INPUT FEATURES AND TARGET

inputs = df[["Pclass", "Sex", "Age", "Fare"]].copy()
target = df["Survived"]

print("\n-----INPUT FEATURES-----")
print(inputs.head())

print("\n-----TARGET-----")
print(target.head())

# STEP 5: LABEL ENCODE THE 'sex' COLUMN

le = LabelEncoder()

inputs["Sex"] = le.fit_transform(inputs["Sex"])

print("\n-----AFTER LABEL ENCODING-----")
print(inputs.head())

# Display the encoding
print("\nEncoding:")
print("Female =", le.transform(["female"])[0])
print("Male =", le.transform(["male"])[0])

# STEP 6: SPLIT TRAINING AND TESTING DATA

inputs_train, inputs_test, target_train, target_test = train_test_split(
    inputs,
    target,
    test_size=0.5,
    random_state=42
)

print("\nTraining Data Shape :", inputs_train.shape)
print("Testing Data Shape :", inputs_test.shape)

print("Training Target Shape :", target_train.shape)
print("Testing Target Shape :", target_test.shape)

# STEP 7: CREATE THE MODEL

model = DecisionTreeClassifier(criterion="gini")

# STEP 8: TRAIN THE MODEL

model.fit(inputs_train, target_train)

# STEP 9: MODEL ACCURACY

train_accuracy = model.score(inputs_train, target_train)
test_accuracy = model.score(inputs_test, target_test)
overall_accuracy = model.score(inputs, target)

print("\n-----ACCURACY-----")
print("Training Accuracy :", train_accuracy)
print("Testing Accuracy  :", test_accuracy)
print("Overall Accuracy  :", overall_accuracy)

# STEP 10: PREDICT NEW PASSENGER

# Pclass = 1
# Sex = Male = 1
# Age = 58
# Fare = 26.5500

prediction = model.predict([[1, 1, 58.0, 26.5500]])

print("\n-----NEW PASSENGER PREDICTION-----")
print("Prediction :", prediction)

if prediction[0] == 1:
    print("The passenger is predicted to Survive.")
else:
    print("The passenger is predicted to Not Survive.")
```
